# Tutorial 2: P22 mouse brain coronal section dataset

## Loading package

```python
import os
import torch
import pandas as pd
import scanpy as sc
```

```python
import SpaLORA
```

```python
# Environment configuration. GPU acceleration is highly recommend for imporoved efficiency.
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

# the location of R. Please replace the path below with local R installation path
os.environ['R_HOME'] = '/opt/R/4.0.3/lib/R'
```

## Loading data

```python
# read data
file_fold = '/root/mouse_brain/' #please replace 'file_fold' with the download path

adata_omics1 = sc.read_h5ad(file_fold + 'mousebrain_rna.h5ad')
adata_omics2 = sc.read_h5ad(file_fold + 'mousebrain_atac.h5ad')

adata_omics1.var_names_make_unique()
adata_omics2.var_names_make_unique()
```

```python
# Specify data type
data_type = 'spatial ATAC–RNA-seq'

# Fix random seed
from SpaLORA.preprocess import fix_seed
random_seed = 2022
fix_seed(random_seed)
```

## Pre-processing data

```python
from SpaLORA.preprocess import clr_normalize_each_cell, pca, lsi

sc.pp.filter_genes(adata_omics1, min_cells=10)
sc.pp.filter_cells(adata_omics1, min_genes=200)
sc.pp.highly_variable_genes(adata_omics1, flavor="seurat_v3", n_top_genes=2000)
sc.pp.normalize_total(adata_omics1, target_sum=1e4)
sc.pp.log1p(adata_omics1)
sc.pp.scale(adata_omics1)

adata_omics1_high =  adata_omics1[:, adata_omics1.var['highly_variable']]
adata_omics1.obsm['feat'] = pca(adata_omics1_high, n_comps=50)
adata_omics1.obsm['raw_feat'] = adata_omics1_high.X.copy()

adata_omics2 = adata_omics2[adata_omics1.obs_names].copy()
if 'X_lsi' not in adata_omics2.obsm.keys():
    sc.pp.highly_variable_genes(adata_omics2, flavor="seurat_v3", n_top_genes=3000)
    lsi(adata_omics2, use_highly_variable=False, n_components=51)
adata_omics2.obsm['feat'] = adata_omics2.obsm['X_lsi'].copy()
```

## Constructing neighbor graph

```python
from SpaLORA.preprocess import construct_neighbor_graph
data = construct_neighbor_graph(adata_omics1, adata_omics2, datatype=data_type)
```

## Training the model

```python
# define model
from SpaLORA.SpaLORA_pyG import Train_SpaLORA
model = Train_SpaLORA(data,datatype=data_type,device=device)

# train model
output = model.train()
```

```python
adata = adata_omics1.copy()
adata.obsm['emb_latent_omics1'] = output['emb_latent_omics1']
adata.obsm['emb_latent_omics2'] = output['emb_latent_omics2']
adata.obsm['SpaLORA'] = output['SpaLORA']
adata.obsm['alpha'] = output['alpha']
adata.obsm['alpha_omics1'] = output['alpha_omics1']
adata.obsm['alpha_omics2'] = output['alpha_omics2']
```

## Cross-omics integrative analysis

```python
from SpaLORA.utils import clustering
clustering(adata, key='SpaLORA', add_key='SpaLORA', n_clusters=9, use_pca=True)
```

```python
# visualization
import matplotlib.pyplot as plt
fig, ax_list = plt.subplots(1, 2, figsize=(12, 5))
sc.pp.neighbors(adata, use_rep='SpaLORA', n_neighbors=30)
sc.tl.umap(adata)

sc.pl.umap(adata, color='SpaLORA', ax=ax_list[0], title='SpaLORA', s=30, show=False)
sc.pl.embedding(adata, basis='spatial', color='SpaLORA', ax=ax_list[1], title='SpaLORA', s=55, show=False)

plt.tight_layout(w_pad=0.3)
plt.show()
```
